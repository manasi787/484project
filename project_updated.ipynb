```python
# write code here
import numpy as np
import pandas as pd
data_frame = pd.read_csv('air-quality-india.csv')
# Convert the timestamp to a datetime object and extract the date
data_frame['Timestamp'] = pd.to_datetime(data_frame['Timestamp'])
data_frame['date'] = data_frame['Timestamp'].dt.date

data_frame.head()
```

```python
# Filter the COVID-19 data for India
data_frame2 = pd.read_csv('owid-covid-data.csv', usecols=["location", "date", "total_cases"])
# Filter the COVID-19 data for India

data_filtered = data_frame2[data_frame2["location"] == "India"].copy()

# Convert date column to datetime object
data_filtered['date'] = pd.to_datetime(data_filtered['date'])

# Filter data for dates up to and including June 2022
data_filtered = data_filtered.loc[data_filtered['date'] <= pd.Timestamp('2022-06-04')]

# Convert date column back to date object if needed
data_filtered['date'] = data_filtered['date'].dt.date

#data_filtered.head()
data_filtered.tail(10)
# Merge the air-quality data with the COVID-19 data on the date column
merged_data = pd.merge(data_frame, data_filtered[['date', 'total_cases']], on='date', how='left')
# Fill 0 in total_case for the dates before the outbreak of COVID-19
merged_data['total_cases'] = merged_data['total_cases'].fillna(0)
merged_data.tail(48)

```
